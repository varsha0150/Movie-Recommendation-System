```python
movies = data[['id','title','genre','overview']]
movies
```

```python
movies['tags']=  movies['overview'] + movies['genre'] 
movies
```

```python
movies_data = movies.drop(columns = ['overview','genre'])
movies_data
```


Applying some ML Algorithms to convert this textual data into vector

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 10000 , stop_words= 'english')
cv
```

```python
vector= cv.fit_transform(movies_data['tags'].values.astype('U')).toarray()
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
 similarity = cosine_similarity(vector)
similarity
```
